- Each box represents the interquartile range (IQR), which includes the middle 50% of the data.
- First Quartile (Q1): This is the value below which 25% of the data falls. It marks the boundary between the lowest 25% of values and the rest. In a salary example, if Q1 is 40,000, it means that 25% of people earn less than 40,000.

- Third Quartile (Q3): This is the value below which 75% of the data falls. It separates the top 25% of values from the rest. If Q3 is 80,000, it means that 75% of people earn less than 80,000, and the remaining 25% earn more than that.

- The line inside the box shows the median salary, which is the middle value of the data for each gender.

-  Each dot represents an individual data point (salary) for a specific gender.
-  The swarm plot arranges the dots so that they do not overlap, giving a sense of the data distribution density.
-  By looking at the dots, you can see where most of the data points are concentrated, as well as the spread and any outliers in the salary distribution.

- Male and Female Salary Distributions: Both genders show a similar spread of salary data, with the median for both genders at about the same level.
- Salary Range: The majority of salaries for both genders fall within the range of 40,000 to 100,000, with a few outliers on the higher and lower ends.
- Density Differences: The concentration of dots within the box and whiskers shows that most salaries are clustered around the median for both genders.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="Gender", y="EstimatedSalary", data=suv_data, showfliers=False)  # Removes outliers for clarity
sns.swarmplot(x="Gender", y="EstimatedSalary", data=suv_data, color=".25")
plt.title("Salary Distribution by Gender (Box + Swarm Plot)")
plt.xlabel("Gender")
plt.ylabel("Estimated Salary")
plt.show()
```

## Use of Get Dummies/ one hot encoding?

```python
suv_data.head()
```

```python
gender_col=pd.get_dummies(suv_data["Gender"],drop_first=True,dtype=int)
gender_col.head()
```

```python
suv_data=pd.concat([suv_data,gender_col],axis=1)
suv_data.head()
```

## Separating X and Y

```python
X=suv_data[["Age","EstimatedSalary","Male"]]
y=suv_data["Purchased"]
```

```python
X.head()
```

```python
y.head()
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2) #0.2 ---> 20 percent of data is for test
```

# Applying Model on Train Data

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg_model=LogisticRegression()
```

```python
log_reg_model.fit(X_train,y_train)
```

```python

y_prediction=log_reg_model.predict(X_test)
y_prediction  #They are like you giving answers in your maths test and y_test value are actual answers from teacher
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_prediction)*100 #marks you score during test
```

#### When you do .reset_index(drop=True), it removes any previous indexing and gives the Series or DataFrame a fresh index, starting from 0. This is useful if your data already has an index that doesnt start from 0 or if the index is not in order

```python
ytest_series = pd.Series(y_test).reset_index(drop=True) #actual values
y_prediction_series = pd.Series(y_prediction).reset_index(drop=True) #they are like your answers

# Concatenate and rename columns
result_df = pd.concat([ytest_series, y_prediction_series], axis=1)
result_df.columns = ['Actual_purchased', 'Predicted_purchased']  # Rename the columns #Comparing Your answers with your teacher's correct anser

result_df
```
